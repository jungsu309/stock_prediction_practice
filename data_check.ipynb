## 한번에 여러 그래프 그리기

#### 시가 고가 저가 종가는 크게 차이를 보이지는 않는듯 보인다.

```python
stock_file.plot(x="일자", y=["시가","고가","저가","종가"], kind = "line", figsize=(10,5))
#거래량과는 단위가 다르기 때문에 y축을 공유할 수 없다 
plt.show()
```

```python
#https://m.blog.naver.com/PostView.naver?isHttpsRedirect=true&blogId=onlyjeje&logNo=221168953266
#plt.rcParams["figure.figsize"] = (10,8)
plt.rcParams["figure.figsize"] = (10,5)
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

line1 = ax1.plot(stock_file["일자"],stock_file["거래량"], color='r', linestyle='--')
line2 = ax2.plot(stock_file["일자"],stock_file["시가"], color='violet', linestyle='--')
line3 = ax2.plot(stock_file["일자"],stock_file["고가"], color='y', linestyle='--' )
line4 = ax2.plot(stock_file["일자"],stock_file["저가"], color='g', linestyle='--' )
line5 = ax2.plot(stock_file["일자"],stock_file["종가"], color='b', linestyle='--')
#line5 = ax2.plot(stock_file["일자"],stock_file["거래량"], color='r', linestyle='--')

ax1.tick_params(axis='y', labelcolor='r')
ax2.tick_params(axis='y', labelcolor='b')

#거래량이 많아서 e로 표현되었는데 그거 풀기
ax1.ticklabel_format(style='plain', axis='y')

fig.tight_layout()
#plt.ticklabel_format(useOffset = False)
#plt.figure(figsize=(10,5))
plt.show()
```

#### 주가가 올라가려고 할 때 거래량이 많이 늘어나는것 같다.

### 값을 전부 정규화 하고 그래프 다시 그려보기

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scale_cols = ['시가', '고가', '저가', '종가', '거래량']
stock_scaled = scaler.fit_transform(stock_file[scale_cols])
stock_scaled = pd.DataFrame(stock_scaled)

```

```python
stock_scaled.columns = ['Open','High','Low','Close','trading value']
stock_scaled.insert(0,'Date', stock_file['일자'])
stock_scaled.head(10)
```

```python
stock_scaled.plot(x="Date", y=['Open','High','Low','Close','trading value'], kind = "line", figsize=(10,5))
plt.show()
```
